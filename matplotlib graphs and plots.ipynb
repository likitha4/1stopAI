```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#1-d array
m=[9,12,10,-1,3]
plt.plot(m)
plt.show()
```

```python
#2-d array
x=[20,14,89,-10]
y=[30,28,77,-8]
plt.plot(x,y)
plt.show()
```

```python
#scatter- points on the graph 
x=[20,14,89,-14,102]
y=[30,28,77,-3,98]
plt.scatter(x,y)
plt.show()
```

```python
##scatter plot
a=50
m=np.random.rand(a)
n=np.random.rand(a)
colors=np.random.rand(a)

circumference=(50*np.pi*np.random.rand(a))  ##50 increases the size of the value point in the plot

plt.scatter(m,n,s=circumference,c=colors,alpha=1)
plt.show()
```

```python
#straight line
slope=-1
const=20
p=np.array([12,18,0,9,3,-4])
q=slope*p+const;
plt.plot(p,q)
plt.show()
```

```python
x=np.array([12,15,18,90])
y=np.array([-3,1,20,40])
plt.plot(x,y,color='green',marker='*')
plt.show()
```

```python
m=([30,10,150,790])
n=([-12,-45,90,87])
plt.plot(m,n,'r--')
plt.show()
```

```python
#2 by 2 matrix 
x=np.random.rand(2,2)
y=np.random.rand(2,2)
print(x,y,sep='\n')
plt.plot(x,y,'m--d')
plt.show()
```

```python
l=np.random.rand(2,3)
k=np.random.rand(2,3)
print(l,k,sep='\n')
plt.plot(k,l,'k^:',c='orange',linewidth=4)#for triangle markers
plt.show()
```

```python
x=np.random.rand(3,3)
y=np.random.rand(3,3)
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.plot(x,y,c='cyan',marker='o')
plt.title('Random values using numpy rand method')
print('Random values using numpy rand method:',x,y,sep='\n')
plt.show()
```

```python
x=np.random.randint(20,size=10)
y=np.random.randint(30,size=10)
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.plot(x,y,c='orange',marker='o')
plt.title('Random values using numpy rand method and using axis method to restrict the values')
plt.axis([2,10,12,20])# xmin,xmax,ymin, ymax
print('Random values using numpy rand method and using axis method to trict the values:',x,y,sep='\n')
```

```python
a=np.random.randint(50,size=20)
b=np.random.randint(30,size=20)
print(a,b,sep='\n')
plt.scatter(a,b)
plt.grid(True)
plt.show()
```

```python
a=np.random.randint(150,size=5)
b=np.random.randint(330,size=5)
print(a,b,sep='\n')
plt.plot(a,b)
plt.xscale('log')
plt.yscale('linear')
plt.show()
```

```python
bookstore='pencils','pens','erasers','notebooks'
sales=[980,3670,120,3900]
plt.pie(sales,explode=(0.025,0.0555,0,0.09),colors=('cyan','red','yellow','orange'),labels=bookstore,autopct='%1.2f%%',shadow=True,startangle=270)
plt.axis('equal')
plt.show()
```

```python
mu=1000
a=np.linspace(500,1000,8)
n,bins,patches=plt.hist(a,50,density=1,facecolor='yellow',alpha=0.65)
```
